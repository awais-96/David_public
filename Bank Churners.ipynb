```
for i in range(len(df1['Marital_Status'])):
  if df1['Marital_Status'] [i] == 'Unknown':
    df1['Marital_Status'] [i] = '0'
  if df1['Marital_Status'] [i] == 'Married':
    df1['Marital_Status'] [i] = '1'
  if df1['Marital_Status'] [i] == 'Divorced':
    df1['Marital_Status'] [i] = '2'
  if df1['Marital_Status'] [i] == 'Single':
    df1['Marital_Status'] [i] = '3'
df1['Marital_Status'] = df1['Marital_Status'].astype(float)
```

```
for i in range(len(df1['Income_Category'])):
  if df1['Income_Category'] [i] == '$120K +':
    df1['Income_Category'] [i] = '0'
  if df1['Income_Category'] [i] == 'Less than $40K':
    df1['Income_Category'] [i] = '1'
  if df1['Income_Category'] [i] == 'Unknown':
    df1['Income_Category'] [i] = '2'
  if df1['Income_Category'] [i] == '$60K - $80K':
    df1['Income_Category'] [i] = '3'
  if df1['Income_Category'] [i] == '$80K - $120K':
    df1['Income_Category'] [i] = '4'
  if df1['Income_Category'] [i] == '$40K - $60K':
    df1['Income_Category'] [i] = '5'
df1['Income_Category'] = df1['Income_Category'].astype(float)
```

```
for i in range(len(df1['Card_Category'])):
  if df1['Card_Category'] [i] == 'Blue':
    df1['Card_Category'] [i] = '0'
  if df1['Card_Category'] [i] == 'Gold':
    df1['Card_Category'] [i] = '1'
  if df1['Card_Category'] [i] == 'Silver':
    df1['Card_Category'] [i] = '2'
  if df1['Card_Category'] [i] == 'Platinum':
    df1['Card_Category'] [i] = '3'
df1['Card_Category'] = df1['Card_Category'].astype(float)
```
